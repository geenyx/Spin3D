```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size": 8})
plt.rc('text', usetex=True)
```

```python
file_path = 'MvsH_Py400nm_EDL030_chip_sat90_me0_000_00001_dat.dat'
field_B, M_EDL030 = np.loadtxt(file_path, delimiter=' ', usecols=(2,3), unpack=True)
```

```python
plt.figure(figsize=(6,4), dpi=150)
plt.plot(field_B/10, M_EDL030*1000, '-')
plt.ylabel('$M_{\mbox{in-plane}}$ [$10^{-3}$ emu]')
plt.xlabel('$B_{90^{\circ}}$ [mT]')
plt.xlim([-100,100])
```
